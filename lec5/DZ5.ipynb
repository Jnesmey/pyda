```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# ПЕРЕЧЕНЬ ПОЛОК

def perechen():
    print('Текущий перечень полок:', end=' ')
    for direct in directories:
        print(f'{int(direct)}' , end=' ' )
        
# ПУНКТ 1 ВЕРНЕМ ИМЯ ПО НОМЕРУ ДОКУМЕНТА КОМАНДА - p

def name_num():
    num = input("Введите номер документа: ")
    for person in documents:
        if person["number"] == num:
            print(f'Владелец документа: {person["name"]}')
            break
    else:
        print("Документ не найден в базе")  
        
# ПУНКТ 2 НА КАКОЙ ПОЛКЕ ХРАНИТСЯ ДОКУМЕНТ - s

def num_dir():
    num = input("Введите номер документа: ")
    for direct in directories:
        if num in directories[direct]:
            print(f'Документ хранится на полке: {direct}')
            break
    else:
         print("Документ не найден в базе")

# ПУНКТ 3 ВСЯ ИНФОРМАЦИЯ О ДОКУМЕНТАХ - l

def inform():
    for person in documents:
        print(f' №: {person["number"]}, тип: {person["type"]}, владелец: {person["name"]}, ', end='')
        for direct in directories:
            if person["number"] in directories[direct]:
                print(f'полка хранения: {direct}')
                
# ПУНКТ 4 ДОБАВИТЬ ПОЛКУ - ads

def ads():    
    num = input('Введите номер полки: ')
    for direct in directories:
        if num == direct:
            print('Такая полка уже существует.')
            break
    else:
        directories.setdefault((num), [])
        print('Полка добавлена.')
    perechen()

# ПУНКТ 5 УДАЛИТЬ ПУСТУЮ ПОЛКУ - ds
def ds():
    num = input('Введите номер полки: ')
    for direct, val in directories.items():
        if direct == num:
            if val != []:
                print('На полке есть документы, удалите из перед удалением полки.')
                perechen()
                break
            else: 
                del(directories[direct])
                print('Полка удалена')
                perechen()
                break
    else:
        print('Такой полки не существует')
```

```python
# ИТОГОВАЯ ПРОГРАММА

comanda = ""
while comanda != "q":
    comanda = input("Введите команду: ")
    if comanda == "p":
        name_num()
    elif comanda == "s":
        num_dir() 
    elif comanda == "l":
        inform() 
    elif comanda == "ads":
        ads() 
    elif comanda == "ds":
        ds() 
```
