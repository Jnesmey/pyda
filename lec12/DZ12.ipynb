```python
import pandas as pd
```

```python
# ЗАДАЧА №1 КЛАССИФИКАТОР ФИЛЬМОВ
# функция для классификации
def classification(row):
    if row['rating'] <= 2:
        return "Низкий рейтинг"
    elif row['rating'] <= 4.0:
        return "Средний рейтинг"
    elif row['rating'] == 5.0 or row['rating'] ==4.5:
        return "Высокий рейтинг"
```

```python
# ЗАДАЧА №1 КЛАССИФИКАТОР ФИЛЬМОВ
# работа функции
df = pd.read_csv('ratings.csv')
df['class'] = df.apply(classification, axis=1)
df.head()
```

```python
# ЗАДАЧА №2 КЛАССИФИКАТОР РЕГИОНОВ
# функция для классификации
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def geo(row):
    for region, cities in geo_data.items():
        for city in cities:
            if city in row['keyword']:
                return region
    else:
        return "undefined"
```

```python
# ЗАДАЧА №2 КЛАССИФИКАТОР РЕГИОНОВ
# работа функции
df2 = pd.read_csv('keywords.csv')
df2['region'] = df2.apply(geo, axis=1)
# проверка корректности работы:
df3 = df2[df2.region != "undefined"]
df3.head()
```

```python
# ЗАДАЧА №3 РАНЬШЕ БЫЛО ЛУЧШЕ
import re
```

```python
# ЗАДАЧА №3 РАНЬШЕ БЫЛО ЛУЧШЕ
# не совсем поняла описанный алгоритм для решения. попыталась сделать через использование регулярных выражений
# вышло не совсем так как надо, но было интересно)))

# функция для классификации:
def years_class(row):
    if re.findall(r'(\d{4})', row['title']):
        return re.findall(r'(\d{4})', row['title'])[0]
    else: 
        return 1900
```

```python
# работа функции
films = pd.read_csv("movies.csv")
films['year'] = films.apply(years_class, axis=1)
raiting = pd.read_csv('ratings.csv')
joined = raiting.merge(films, on='movieId', how='left')
years_rating = joined.groupby('year').mean()
years_rating.sort_values(by = 'rating', ascending = False).head()

# таким способом часть годов перепутается, если фильм назван четыремя цифрами
```
