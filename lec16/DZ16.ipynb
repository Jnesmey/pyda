```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
# Взяты данные о уровне счастья по странам мира за 2019 год
# Датасет: - страна, оценка, ВВП, социальная поддержка, Ожидаемость здоровой жизни, свобода выбора, щедрость коррупция
df = pd.read_csv('2019.csv')
df.head()
```

```python
df.describe()
```

```python
# Центральный показатель - уровень счастья. Посмотрим на его распределение с помощью боксплота и гистограммы
df[['Score']].plot(kind='box', title='Рейтинг счастья')
# Распределение достаточно равномерное, выборосов нет. 
# Ящик имеет "ровные усы". С реднем рейтинг счастья находится в интервале от 4.5 до 6
```

```python
df_score = df.Score.plot(kind='hist', bins = 20)
df_score.set_xlabel('Оценка')
df_score.set_ylabel('Частота')
df_score.set_title('Распределение стран по уровню счастья')

# Большинство стран имеет средний уровень счастья. Рейтинг ниже 4 и больше 7.5 встречается гораздо реже                    
```

```python
# Посмотрим взаимосвязь показателей влияющих на рейтинг
sns.pairplot(df, corner=True)
# По графикам предполагаем, что большее влияние на счастье оказывает ВВП, социальная помощь и здоровая жизь
```

```python
# Рассмотрим как распределенны страны по показателям
sns.boxplot(data=df[['GDP per capita','Social support', 'Healthy life expectancy']])
```

```python
# Рассмотрим как распределенны страны по показателям
sns.boxplot(data=df[['Freedom to make life choices','Generosity','Perceptions of corruption']])
```

```python
# Взаимосвязь показателей

a1 = df.plot(kind='scatter',
        x='GDP per capita',
        y='Social support',
        s=df['Score']*50)
a1.set_title('Корреляция между уровенем счастья, ВВП и социальной поддержкой')

# Уровень счастья растет при увеличении ВВП и социальной поддержки
```

```python
# Взаимосвязь показателей
a2 = df.plot(kind='scatter',
        x='Healthy life expectancy',
        y='Social support',
        s=df['Score']*50)
a1.set_title('Корреляция между уровенем счастья, здоровой жизнью и социальной поддержкой')

# Уровень счастья растет при увеличении здоровой жизни и социальной поддержки
```
