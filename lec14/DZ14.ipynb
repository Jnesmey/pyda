```python
# # Задание 1

ratings = pd.read_csv('ratings.csv')
df = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'rating': 'count'})
df['time_life'] = df['timestamp']['max'] - df['timestamp']['min']
result = df[df['rating']['count'] > 100]
mean_time_life = result['time_life'].mean()
mean_time_life
```
